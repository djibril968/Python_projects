```python
import numpy as np
import pandas as pd
```

```python
#data cleaning
#here we carry out the following operations

"""remove duplicates
look out for missing number"""
```

```python

hr_data_01 = pd.read_csv('./HR_data_01.csv')
hr_data_01.head()
```

```python
#looking out for duplicates

hr_data_dup = hr_data_01.duplicated()
hr_data_dup

#our output shows we have no duplicates in our dataset
```

```python
#checking through for missing values

hr_data_missing = hr_data_01.isnull()
hr_data_missing
```

```python
#Here we look at our columns

hr_data_cols = hr_data_01.columns
hr_data_cols
```

```python
#here we commence our eda

"counts of attrition by department"

att_by_dept = hr_data_01.groupby("Department")['Attrition'].value_counts().reset_index(name = 'counts').sort_values(['Department', 'counts'], ascending=([True, False]))
att_by_dept

```

```python
#total employees by department

emp_by_dept = hr_data_01.groupby('Department')['emp no'].count().sort_values(ascending= True)
emp_by_dept
```

```python
#total satisfied by attrition
sat_att = hr_data_01.groupby(['Department','Job Satisfaction'])['Attrition'].value_counts(normalize= True)
sat_att
```

```python
sat_att_norm = hr_data_01.groupby(['Department','Job Satisfaction'])['Attrition'].value_counts()
sat_att_norm
```

```python
#average rating by department
def avg(column):
    return column.mean

dept_ratings = hr_data_01.groupby('Department')['Performance Rating'].agg(np.mean)
dept_ratings
```
