```python
#now we shall expand our fact table to include the following columns
#this allows us get the weekly email volume
"""
day gives us the day number in the month
day_name = this gives us the name of the day in the week, 
week_no"""
#day
Email_fact_data['Day_of_mon'] = Email_fact_data['Msg_Date'].dt.day

#day_name

Email_fact_data['day_name'] = Email_fact_data['Msg_Date'].dt.day_name()
Email_fact_data
```

```python
#next we shall create a custom funcdtion to enable us get the week no

#first we define our function
def week_of_month(dt):
    #next we extract the first day of the month
    first_day = dt.replace(day = 1)
    dom = dt.day
    adjusted_dom = first_day.weekday() + dom
    return  int(np.ceil(adjusted_dom/7.0))



#week_no
week_extract = Email_fact_data['Msg_Date'].apply(week_of_month)
Email_fact_data['Week_of_month'] = 'week' + ' ' + week_extract.astype(str)
Email_fact_data
```

```python
#email exchange type
#here we come up with two categories for inter-dept and intra_dept mails that helps us
email_exc_dept['exc_type'] = ''

for i,  (Sender_dept, Rec_dept) in enumerate(zip(email_exc_dept['Sender_dept'], email_exc_dept['Rec_dept'])):
    if Sender_dept == Rec_dept:
        email_exc_dept.at[i,'exc_type'] = 'intra_mails'
    else:
        email_exc_dept.at[i,'exc_type'] = 'inter_mail'
         
```
