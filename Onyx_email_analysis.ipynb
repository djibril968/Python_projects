```python
#Is there any signifcance between the period when messages were sent and their status (using chisquare)

import scipy
from scipy.stats import chi2_contingency

chi, p_val, df, expected = scipy.stats.chi2_contingency(Test_01)
chi, p_val, df, expected

result_test_01 = pd.DataFrame({'chi' : [chi],
                               'p_val' : [p_val],
                                'df': [df] })

result_test_01
```

```python
#now we use regression model to predict the likelihood of message_status by workday
#what are the chances that a mail will be opened if it was sent within workday

#first we import our regression library

import statsmodels as sm
import statsmodels.api as ssm
```

```python
#due to the nature of our data (categorical data), we will have to encode values from each columns

Email_fact_data['Msg_stat_coded'] = Email_fact_data['Message_status'].map({'opened' : 1, 'unopened': 0})
Email_fact_data['Within_wd'] = Email_fact_data['Within_workdays'].map({'yes': 1, 'no':0})

#now we pass the new columns to our x, y variables
x = Email_fact_data['Within_wd']
y = Email_fact_data['Msg_stat_coded']

#now we add a consant

X = ssm.add_constant(x)

#now we call the logistic regression function
reg_model = ssm.Logit(y, X).fit()

print(reg_model.summary())
```

```python
#now we plot the result of our model   first we get our predictor value
Email_fact_data['pred_prob'] = reg_model.predict(X)
#next we go on to import our library
import matplotlib.pyplot as plt

#next we pull the actual data
plt.scatter(Email_fact_data['Within_wd'], Email_fact_data['Message_status'], c = 'blue', label = 'actual data')

#now we plot our logistice regression probability
plt.scatter(Email_fact_data['Within_wd'], Email_fact_data['pred_prob'], c = 'red', label = 'predicted probability', alpha= 0.5)
plt.xlabel('Mail within workdays')
plt.ylabel('Message status')
plt.title('Predicted probabilities')
plt.show()
```

```python
#now we shall look into how mails within work hours influence message status

#we shall utilize chisquare to measure associations
#logistic regression to predict values
#scatter plot to display result

#now we pivot our data for use
work_hrs_piv = Email_fact_data.pivot_table(index= 'Within_work_hrs', columns= 'Message_status',  aggfunc= 'size')
work_hrs_piv

#now we introduce our chi_square function

chi, p_val, df, expected = scipy.stats.chi2_contingency(work_hrs_piv)

#now we pass our result intp a dataframe
whrs_chi = pd.DataFrame({'chi': [chi],
                         'p_val' : [p_val],
                         'df': [df]})
whrs_chi
```

```python
#Now we move on to our logistic regression analysis

#first we encode the values from our columns and pass them into new columns

Email_fact_data['Within_hrs'] = Email_fact_data['Within_work_hrs'].map({'yes' : 1, 'no' : 0})

#now we pass the required columns into our x, y varables

x = Email_fact_data['Within_hrs']
y = Email_fact_data['Msg_stat_coded']

#now we add constant

X = ssm.add_constant(x)

#now we call our logistic regression function
reg_model_01 = ssm.Logit(y, X).fit()

print(reg_model_01.summary())
```

```python
#logistic regression to predict if email subject influences the massage status
#first we encode our variables
Email_fact_data['Topic_coded'] = Email_fact_data['Email_subject'].map({'Sales and Marketing' : 1, 'Legal Matters' : 2,
                                                                       'Meeting Scheduling' : 3, 'Operational Issues': 4,
                                                                      'Project Updates' : 5, 'Technical Support' : 6,
                                                                       'Product Development': 7, 'Internal Policies and Procedures' : 8,
                                                                        'Client Related Communication': 9, 'HR related topics': 10,
                                                                         'Finance and Accounting': 11, 'Training and Development' : 12,
                                                                          'Vendor Relations' : 13, 'Security Alerts' : 14 })
#next we pass data into our x, y variables

x = Email_fact_data['Topic_coded']
y = Email_fact_data['Msg_stat_coded']

#next we add our constant
X = ssm.add_constant(x)

#next we call our logistic regression function
reg_model_02 = ssm.Logit(y, X).fit()

#print result
print(reg_model_02.summary())
```
