```python
#departments with most communication for each week

dept_com_by_week = Email_fact_data.groupby(['Week_of_month', 'Sender_dept'])['Rec_dept'].value_counts().reset_index(name= 'mail_count')
dept_com_by_week

dept_comm_by_week = pd.DataFrame(columns= ['Week_of_month', 'Sender_dept', 'Rec_dept', 'mail_count'])

#now we index 
dept_com_idx = dept_com_by_week.groupby(['Week_of_month', 'Sender_dept'])['mail_count'].idxmax()

#now we pass our indexed df into dept_comm_by_week
dept_comm_by_week = dept_com_by_week.loc[dept_com_idx].reset_index(drop= True)
dept_comm_by_week
```

```python
# Step 1: Group by week, sender department, and recipient department, then count emails
dept_com_by_week = Email_fact_data.groupby(['Week_of_month', 'Sender_dept', 'Rec_dept']).size().reset_index(name='mail_count')

# Step 2: Find the recipient department with the maximum email count for each week and sender department
dept_com_idx = dept_com_by_week.loc[dept_com_by_week.groupby(['Week_of_month', 'Sender_dept'])['mail_count'].idxmax()]

# Step 3: Reset index for the final DataFrame
dept_comm_by_week = dept_com_idx.reset_index(drop=True)

# Display the result
dept_comm_by_week
```
