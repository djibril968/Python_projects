```python
#volume of email exchnage between departments
email_exc_dept = Email_fact_data.groupby('Sender_dept')['Rec_dept'].value_counts().reset_index(name = "email_vol").sort_values('email_vol', ascending = False)
email_exc_dept

#pd.set_option('display.max_row', 39)
```

```python
#intra departmental mails
email_exc_dept['exc_type'] = ''

for i,  (Sender_dept, Rec_dept) in enumerate(zip(email_exc_dept['Sender_dept'], email_exc_dept['Rec_dept'])):
    if Sender_dept == Rec_dept:
        email_exc_dept.at[i,'exc_type'] = 'intra_mails'
    else:
        email_exc_dept.at[i,'exc_type'] = 'inter_mail'



      
        
```

```python
#expanded email fact data

Email_fact_data['mail_exc_type'] = ''


for i, (Sender_dept, Rec_dept) in enumerate(zip(Email_fact_data['Sender_dept'],Email_fact_data['Rec_dept'])):
    if Sender_dept == Rec_dept:
        Email_fact_data.at[1, 'mail_exc_type'] = 'intra_mails'
    else:
        Email_fact_data.at[1, 'mail_exc_type'] = 'inter_mails'
```
